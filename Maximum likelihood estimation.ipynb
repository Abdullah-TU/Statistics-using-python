```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm


# calculating pdf for values of x in between -6 to 6


x= np.linspace(start=-6, stop=6, num=200) #Return evenly spaced numbers over a specified interval.

plt.figure(num=0, figsize=(10,8))


# A normal continuous random variable.
#The location (loc) keyword specifies the mean. 
#The scale (scale) keyword specifies the standard deviation.
#Specifically, norm.pdf(x, loc, scale) is identically equivalent to norm.pdf(y) / scale with y = (x - loc) / scale.

plt.plot(x, norm.pdf(x, loc=0, scale=1), color="green", label ="$\\mu=0$, $\sigma=1$")
plt.plot(x, norm.pdf(x, loc=1, scale=1.5), color="red", label ="$\\mu=0$, $\sigma=1$")
plt.plot(x, norm.pdf(x, loc=-1, scale=0.5), color="blue", label ="$\\mu=0$, $\sigma=1$")
plt.plot(x, norm.pdf(x, loc=1.5, scale=2), color="violet", label ="$\\mu=0$, $\sigma=1$")

plt.title("Probabbility desnsity function of $N(\\mu, \\sigma^2)$")
plt.ylabel("Density")

plt.legend()
plt.show()
```

```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(19680801)

# example data
mu = -1  # mean of distribution
sigma = 0.5  # standard deviation of distribution
x = mu + sigma * np.random.randn(437)

num_bins = 50  # bin means number of blocks try num_bins = 2, There are several different terms for binning including bucketing, discrete binning, discretization or quantization

fig, ax = plt.subplots()

# the histogram of the data
n, bins, patches = ax.hist(x, num_bins, density=True)

# add a 'best fit' line
y = ((1 / (np.sqrt(2 * np.pi) * sigma)) *
     np.exp(-0.5 * (1 / sigma * (bins - mu))**2))
ax.plot(bins, y, '-', color="red")
ax.set_xlabel('Smarts')
ax.set_ylabel('Probability density')
ax.set_title(r'Histogram of IQ: $\mu=100$, $\sigma=15$')

# Tweak spacing to prevent clipping of ylabel
fig.tight_layout()
plt.show()
```

```python
from scipy.stats import chi2
import matplotlib.pyplot as plt

x=np.linspace(0,10,200)


plt.figure(figsize=(10,8))


plt.plot(x, chi2.pdf(x, df=1), color="red", label="k=1")



for i in range(2,11):
    plt.plot(x, chi2.pdf(x, df=i), label=" k ="+str(i))

plt.ylim((0, 0.6))    
plt.title("Probability density function of $\\chi^2_k$")
plt.ylabel("Density")
plt.legend()
    
plt.show()
```

```python
from scipy.stats import t

x=np.linspace(-5,5,200)


plt.figure (figsize=(10,8))

m=[1,2,4,5,6,7,10,12,15,16]
for i in range(0,10):
    plt.plot(x, t.pdf(x, df=m[i]),
            label = "k = " + str(m[i]))
    
plt.ylim((0, 0.4))    
plt.title("Probability density function of $t_k$")
plt.ylabel("Density")
plt.legend()
plt.show()
```

```python
from scipy.stats import f

x= np.linspace(1e-10, 3, 200)

plt.figure(num = 3, figsize = (10, 8))

plt.plot(x, f.pdf(x, dfn = 1, dfd = 1), label = "$k_1 = 1$, $k_2 = 1$")
plt.plot(x, f.pdf(x, dfn = 3, dfd = 2), label = "$k_1 = 3$, $k_2 = 2$")
plt.plot(x, f.pdf(x, dfn = 2, dfd = 3), 
         linestyle = "--", label = "$k_1 = 2$, $k_2 = 3$")
plt.plot(x, f.pdf(x, dfn = 6, dfd = 3), label = "$k_1 = 6$, $k_2 = 3$")
plt.plot(x, f.pdf(x, dfn = 3, dfd = 6), 
         linestyle = "-", label = "$k_1 = 3$, $k_2 = 6$")
plt.plot(x, f.pdf(x, dfn = 100, dfd = 100), label = "$k_1 = 100$, $k_2 = 100$")


plt.ylim((0, 2.1))    
plt.title("Probability density function of $F_{k_1, k_2}$")
plt.ylabel("Density")

plt.legend()
plt.show()
```
