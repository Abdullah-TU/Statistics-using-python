# Normal Distribution

```python
# importing libraries for normal distribution calculation

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm


# calculating the probability density function for values( x in [-10, 10])
x= np.linspace(start = -5, stop = 5, num = 200)

plt.figure(num = 0, figsize = (10, 8))

plt.plot(x, norm.pdf(x, loc = 0, scale = 1), color = "red", 
         label = "$\\mu=0$, $\sigma =1$")

plt.title("Probability density function of $N(\\mu, \\sigma^2)$")
plt.ylabel("Density")
plt.legend()
plt.show()


```

# chi-squared distribution




```python
from scipy.stats import chi2

x= np.linspace(start=0, stop =10, num =200)
    
plt.figure(figsize=(10,8))

plt.plot(x, chi2.pdf(x, df=1))

for i in range(2,7):
    plt.plot(x, chi2.pdf(x, df=i))
plt.show()
```

#  t-Distribution

```python
from scipy.stats import t

from scipy.stats import chi2

x= np.linspace(start=-6, stop =6, num =200)
    
plt.figure(figsize=(10,8))

plt.plot(x, t.pdf(x, df=1))

for i in range(2,7):
    plt.plot(x, t.pdf(x, df=i))
plt.show()
```

```python
from scipy.stats import t

from scipy.stats import chi2

x= np.linspace(start=-6, stop =6, num =200)
    
plt.figure(figsize=(10,8))

plt.plot(x, t.pdf(x, df=1))

for i in range(0,5):
    plt.plot(x, t.pdf(x, df=[1, 2, 3, 5, 100][i]))
plt.show()
```

# F-Distribution

```python
from scipy.stats import f

x= np.linspace(start=-4, stop =4, num =200)
    
plt.figure(figsize=(10,8))

plt.plot(x, f.pdf(x, dfn = 1, dfd = 1))
plt.plot(x, f.pdf(x, dfn = 3, dfd = 2))
plt.plot(x, f.pdf(x, dfn = 2, dfd = 3))
plt.plot(x, f.pdf(x, dfn = 3, dfd = 6))
plt.plot(x, f.pdf(x, dfn = 100, dfd = 100))         
         
plt.show()
```

# Log-Normal Distribution

```python
from scipy.stats import lognorm

x= np.linspace(start=-4, stop =4, num =200)
    
plt.figure(figsize=(10,8))

plt.plot(x, lognorm.pdf(x, s=0.954, loc=2, scale=0.954))
     
         
plt.show()
```

# Gamma Distribution

```python
from scipy.stats import gamma

x= np.linspace(start=-4, stop =4, num =200)
    
plt.figure(figsize=(10,8))

plt.plot(x, gamma.pdf(x, a = 1.99323054838, loc=0, scale=1))
     
         
plt.show()
```
